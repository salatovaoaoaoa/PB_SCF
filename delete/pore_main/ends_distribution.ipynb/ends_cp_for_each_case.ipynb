```python
import sys
import os
sys.path.append('/home/tpopova/prj/PB_SCF')

from pore_main.pore_utils import pore_utils
from py_utils.generate_pore_in_files import generate_pore_in_files
from py_utils.generate_flat_cap_in_files import generate_flat_cap_in_files
from py_utils.generate_two_flat_brushes import generate_two_flat_brush_in_files


import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
N = 300
S = 100
Cs = 0.001
alpha = 0.5
N_layers = np.linspace(80, 80, num = 1)
```

Surface (одна щетка с крышкой)

```python
colors = ['fuchsia', 'blue', 'limegreen', 'orangered', 'orange', 'skyblue', 'slateblue', 'black', 'hotpink',
              'crimson', 'aquamarine']

fig = plt.figure()

ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

for n, color in zip(N_layers, colors):
    file_name_pro_flat_cap_surface = generate_flat_cap_in_files(template_cap = '/home/tpopova/prj/PB_SCF/pore_templates/temp_surface.in',
                    
                    #название параметра
                    range_param = 'N_layers',
                    
                    type_lowerbound = 'surface',
                    
                    N_brush = N,
                    N_layers = round(n),
                    S = S,
                    Cs = Cs,
                    alpha = alpha,
                    min_range_value = 4,
                    max_range_value = 4,
                    )

    c_polymer_surface = np.asarray(pd.read_csv(file_name_pro_flat_cap_surface, sep='\t')['mol_brush_phi'])

    ends_surface = np.asarray(pd.read_csv(file_name_pro_flat_cap_surface, sep='\t')['mon_G_phi'])
    
    z_range = np.linspace(0, round(n), num = round(n)+2)
    
        
    ax1.plot(z_range, c_polymer_surface,
                linestyle='-',
                linewidth= 2,
                color=color,
                label=f'$N_{{layers}} = {round(n)}$')
    
#     ax1.plot(-1 * z_range, c_polymer_surface,
#                 linestyle='-',
#                 linewidth= 2,
#                 color=color)
    
    ax2.plot(z_range, ends_surface,
            linestyle='-',
            linewidth= 2,
            color=color,
            label=f'$N_{{layers}} = {round(n)}$')
    
#     ax2.plot(-1 * z_range, ends_surface,
#             linestyle='-',
#             linewidth= 2,
#             color=color)

#  Добавляем подписи к осям:
ax1.set_xlabel('z',
                fontsize=23,  
                color='black', labelpad=2.5) 

ax1.set_ylabel(r'$\phi(z)$',
                fontsize=21, 
                color='black')
ax2.set_xlabel('z',
                fontsize=23, 
                color='black') 

ax2.set_ylabel('$ends$',
                fontsize=21,
                color='black') 

# legend
ax1.legend(title = f'$boundary = surface, N = {N}, S = {S}, Cs = {Cs}$',
    fontsize=8,
            ncol=1,  # количество столбцов
            facecolor='white',  # цвет области
            edgecolor='black',  # цвет крайней линии
            title_fontsize='8',  # размер шрифта заголовка
            loc = 'center right')
# legend
ax2.legend(fontsize=8,
            ncol=1,  # количество столбцов
            facecolor='white',  # цвет области
            edgecolor='black',  # цвет крайней линии   #  заголовок
            title_fontsize='8', loc = 'upper right'  # размер шрифта заголовка
            )

fig.set_figwidth(9)
fig.set_figheight(11)

plt.rcParams['xtick.labelsize'] = 20
plt.rcParams['ytick.labelsize'] = 20

plt.show()
```

Щетка с крышкой, mirror

```python
N_layers = np.linspace(80, 80, num = 1)
```

```python
colors = ['fuchsia', 'blue', 'limegreen', 'orangered', 'orange', 'skyblue', 'slateblue', 'black', 'hotpink',
              'crimson', 'aquamarine']

fig = plt.figure()

ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

for n, color in zip(N_layers, colors):
    file_name_pro_flat_cap_mirror = generate_flat_cap_in_files(template_cap = '/home/tpopova/prj/PB_SCF/pore_templates/temp_surface.in',
                    
                    #название параметра
                    range_param = 'N_layers',
                    
                    type_lowerbound = 'mirror',
                    
                    N_brush = N,
                    N_layers = round(n),
                    S = S,
                    Cs = Cs,
                    alpha = alpha,
                    min_range_value = min(N_layers),
                    max_range_value = max(N_layers),
                    )

    c_polymer_mirror = np.asarray(pd.read_csv(file_name_pro_flat_cap_mirror, sep='\t')['mol_brush_phi'])

    ends_mirror = np.asarray(pd.read_csv(file_name_pro_flat_cap_mirror, sep='\t')['mon_G_phi'])
    
    z_range = np.linspace(0, round(n), num = round(n)+2)
    
    ax1.plot(z_range, c_polymer_mirror,
                linestyle='-',
                linewidth= 2,
                color=color,
                label=f'$N_{{layers}} = {round(n)}$')
    
    ax1.plot(-1 * z_range, c_polymer_mirror,
                linestyle='-',
                linewidth= 2,
                color=color)
    
    ax2.plot(z_range, ends_mirror,
            linestyle='-',
            linewidth= 2,
            color=color,
            label=f'$N_{{layers}} = {round(n)}$')
    
    ax2.plot(-1 * z_range, ends_mirror,
            linestyle='-',
            linewidth= 2,
            color=color)

#  Добавляем подписи к осям:
ax1.set_xlabel('z',
                fontsize=23,  
                color='black', labelpad=2.5) 

ax1.set_ylabel(r'$\phi(z)$',
                fontsize=21, 
                color='black')
ax2.set_xlabel('z',
                fontsize=23, 
                color='black') 

ax2.set_ylabel('$ends$',
                fontsize=21,
                color='black') 

# legend
ax1.legend(title = f'$boundary = mirror, N = {N}, S = {S}, Cs = {Cs}$',
    fontsize=8,
            ncol=1,  # количество столбцов
            facecolor='white',  # цвет области
            edgecolor='black',  # цвет крайней линии
            title_fontsize='8',  # размер шрифта заголовка
            loc = 'upper right')
# legend
ax2.legend(fontsize=8,
            ncol=1,  # количество столбцов
            facecolor='white',  # цвет области
            edgecolor='black',  # цвет крайней линии   #  заголовок
            title_fontsize='8', loc = 'upper left'  # размер шрифта заголовка
            )

fig.set_figwidth(9)
fig.set_figheight(11)

plt.rcParams['xtick.labelsize'] = 20
plt.rcParams['ytick.labelsize'] = 20

plt.show()
```

```python
colors = ['fuchsia', 'blue', 'limegreen', 'orangered', 'orange', 'skyblue', 'slateblue', 'black', 'hotpink',
              'crimson', 'aquamarine']

fig = plt.figure()

ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

for n, color in zip(N_layers, colors):
    file_name_pro_two_brushes = generate_two_flat_brush_in_files(template_two = '/home/tpopova/prj/PB_SCF/pore_templates/two_flat_compress_temp.in',
                   
                   #название параметра
                   range_param = 'N_layers',
                   
                   N_brush_left = N,
                   N_brush_right = N,
                   N_layers = round(n)*2,
                   S_left = S,
                   S_right = S,
                   Cs = Cs,
                   alpha_left =alpha,
                   alpha_right = alpha,
                   min_range_value = min(N_layers),
                   max_range_value = max(N_layers),
                   )

    c_polymer_right = np.asarray(pd.read_csv(file_name_pro_two_brushes, sep='\t')['mol_brushright_phi'])
    c_polymer_left = np.asarray(pd.read_csv(file_name_pro_two_brushes, sep='\t')['mol_brushleft_phi'])

    ends_right = np.asarray(pd.read_csv(file_name_pro_two_brushes, sep='\t')['mon_E_phi'])
    ends_left = np.asarray(pd.read_csv(file_name_pro_two_brushes, sep='\t')['mon_G_phi'])
    
        
    ax1.plot(c_polymer_right,
                linestyle='-',
                linewidth= 2,
                color=color,
                label=f'$N_{{layers}} = {round(n)}$')
    
    ax1.plot(c_polymer_left,
                linestyle='-',
                linewidth= 2,
                color=color)
    
    ax2.plot(ends_right,
            linestyle='-',
            linewidth= 2,
            color=color,
            label=f'$N_{{layers}} = {round(n)}$')
    
    ax2.plot(ends_left,
            linestyle='-',
            linewidth= 2,
            color=color)

#  Добавляем подписи к осям:
ax1.set_xlabel('z',
                fontsize=23,  
                color='black', labelpad=2.5) 

ax1.set_ylabel(r'$\phi(z)$',
                fontsize=21, 
                color='black')
ax2.set_xlabel('z',
                fontsize=23, 
                color='black') 

ax2.set_ylabel('$ends$',
                fontsize=21,
                color='black') 

# legend
ax1.legend(title = f'$two \\ brushes, N = {N}, S = {S}, Cs = {Cs}$',
    fontsize=8,
            ncol=1,  # количество столбцов
            facecolor='white',  # цвет области
            edgecolor='black',  # цвет крайней линии
            title_fontsize='8',  # размер шрифта заголовка
            loc = 'upper right')
# legend
ax2.legend(fontsize=8,
            ncol=1,  # количество столбцов
            facecolor='white',  # цвет области
            edgecolor='black',  # цвет крайней линии   #  заголовок
            title_fontsize='8', loc = 'upper right'  # размер шрифта заголовка
            )

fig.set_figwidth(9)
fig.set_figheight(11)

plt.rcParams['xtick.labelsize'] = 20
plt.rcParams['ytick.labelsize'] = 20

plt.show()
```
